```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0  # Normalize to [0,1]
```

```python
def create_model():
    model = Sequential([
        Flatten(input_shape=(28, 28)),
        Dense(128, activation='relu'),
        Dense(10, activation='softmax')
    ])
    model.compile(optimizer=Adam(),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model

quantization_levels = [8, 16, 32, 64]
```

```python
# Quantization of ReLu
def quantize_activation(x, levels):
    """Quantize activation outputs to a fixed number of levels."""
#     x = tf.clip_by_value(x, 0, 1)  
    min_val = tf.reduce_min(x)
    max_val = tf.reduce_max(x)
    x = (x - min_val) / (max_val - min_val) # Normalize x to [0, 1]
    return tf.round(x * (levels - 1)) / (levels - 1)

   
# Create and train the original model
model = create_model()
history = model.fit(x_train, y_train, epochs=5, validation_split=0.1)
train_accuracy = history.history['accuracy'][-1]
print(f"Training Accuracy for original: {train_accuracy:.4f}")
    
histories = []

for levels in quantization_levels:
    # Create a new model with quantized activation
    model_quantized = create_model()
    for layer in model_quantized.layers:
        if hasattr(layer, 'activation') and layer.activation == tf.nn.relu:  # quantize the relu
            layer.activation = lambda x: quantize_activation(x, levels=levels)
            
    # Compile the model
    model_quantized.compile(optimizer=Adam(),
                            loss='sparse_categorical_crossentropy',
                            metrics=['accuracy'])
    
    # Train the model
    print(f"Training with quantization levels: {levels}")
    history_quantized = model_quantized.fit(x_train, y_train, epochs=5, validation_split=0.1, verbose=0)
    histories.append(history_quantized)
    train_accuracy = history_quantized.history['accuracy'][-1]
    print(f"Training Accuracy: {train_accuracy:.4f}")
    
    
plt.figure(figsize=(15, 9))
# Plot original model accuracy
plt.plot(history.history['accuracy'], label='Original (No Quantization)')
# Plot quantized model accuracies
for levels, history in zip(quantization_levels, histories):
    plt.plot(history.history['accuracy'], label=f'Quantized ReLu({levels} levels)')

plt.title('Model Accuracy with quantized ReLu')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
